```python
			#import library
import numpy as np
import os
import math
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
```

```python
			#Taking picture

webcam =cv2.VideoCapture(0)
ret, frame=webcam.read()
print (ret)
webcam.release()  #close the webcam
```

```python
	#opening in window
cv2.startWindowThread()
cv2.namedWindow("Picture Window",cv2.WINDOW_NORMAL)
cv2.imshow("Picture Window",frame)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#open in window and print in browser

cv2.namedWindow("Picture Window",cv2.WINDOW_AUTOSIZE)
cv2.imshow("Picture Window",frame)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(frame)
plt.show()
```

```python
# BGR --->>>> RGB
#opencv takes picture in BGR.For better quality we have to convert it to RGB

frame_RGB = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(frame_RGB)
plt.axis("off")
plt.show()
```
